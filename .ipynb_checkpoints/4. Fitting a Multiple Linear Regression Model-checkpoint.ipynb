```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Let's read the data...
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
#..and create a new data frame in case we want later on
# to work with the initial dataframe
train = train_data.copy()
test = test_data.copy()
```

```python
# From the EDA notebook we know that 'GrLivArea', 'GarageArea', 'TotalBsmtSF' are the 
# quantitative variablews with the highest correlation with our target variable 'SalePrice'. 
# So let's use them to build a multiple linear regression model.
columns = ['GrLivArea', 'GarageArea', 'TotalBsmtSF', 'SalePrice']
# Creating a new dataframe which contains only the target variable and
# the selected predictor variable
play_dataset = pd.DataFrame(train, columns=columns)
# Checking correlation of the selected variables with our target varaible once again:
print(round(play_dataset['GrLivArea'].corr(play_dataset['SalePrice']), 3))
print(round(play_dataset['GarageArea'].corr(play_dataset['SalePrice']), 3))
print(round(play_dataset['TotalBsmtSF'].corr(play_dataset['SalePrice']), 3))
```

```python
# Plotting the scatterplots of the selected variables vs the target variable:
```

```python
plt.scatter(play_dataset['GrLivArea'], play_dataset['SalePrice'])
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.show()
```

```python
plt.scatter(play_dataset['GarageArea'], play_dataset['SalePrice'])
plt.xlabel('GarageArea')
plt.ylabel('SalePrice')
plt.show()
```

```python
plt.scatter(play_dataset['TotalBsmtSF'], play_dataset['SalePrice'])
plt.xlabel('TotalBsmtSF')
plt.ylabel('SalePrice')
plt.show()
```

```python
# Let's select a small part of the dataset for educational purposes: 
small_dataset = play_dataset[:20]
```

```python
# Correlations still relatively high:
print(round(play_dataset['GrLivArea'].corr(small_dataset['SalePrice']), 3))
print(round(play_dataset['GarageArea'].corr(small_dataset['SalePrice']), 3))
print(round(play_dataset['TotalBsmtSF'].corr(small_dataset['SalePrice']), 3))
```

```python
# Plotting the new scatterplots of the small dataset:
```

```python
plt.scatter(small_dataset['GrLivArea'], small_dataset['SalePrice'])
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.show()
```

```python
plt.scatter(small_dataset['GarageArea'], small_dataset['SalePrice'])
plt.xlabel('GarageArea')
plt.ylabel('SalePrice')
plt.show()
```

```python
plt.scatter(small_dataset['TotalBsmtSF'], small_dataset['SalePrice'])
plt.xlabel('TotalBsmtSF')
plt.ylabel('SalePrice')
plt.show()
```

```python
# Lets fit a regression line to the above small dataset:
from sklearn.linear_model import LinearRegression
```

```python
X = small_dataset[['GrLivArea', 'GarageArea', 'TotalBsmtSF']]
y = small_dataset['SalePrice']
```

```python
# We create the regressor lm which will apply the Ordinary least squares (OLS) 
# method in order to fit a linear regression line to our our small_dataset samples
lm = LinearRegression()
lm.fit(X, y)
# Printing the estimated coefficient for the X variable: 
print(lm.coef_)
# Printing the intercept of the model:
print(lm.intercept_)
```

```python
# Just manually checking how the predict method works 
print(round(lm.predict(5)))
print(round(24290.63628644 + 107.15314929 * 5))
```

```python
# Let's see how the coefficient and the intercept 
# change if we feed all data to the regressor
```

```python
X_all = train[['GrLivArea', 'GarageArea', 'TotalBsmtSF']]
y_all = train['SalePrice']
lm_all = LinearRegression()
lm_all.fit(X_all, y_all)
# Printing the estimated coefficient for the X variable: 
print(lm_all.coef_)
# Printing the intercept of the model:
print(lm_all.intercept_)
```

```python
# The coefficients and the intercept change a lot in this case 
# when we use all available data
```

```python
# Lets check how close to real y values are
# the predicted y values. We will calculate
# the root mean square error (RMSE)
# in the training data set (we didnt use a test set)
# Using the small dataset:
from sklearn.metrics import mean_squared_error
rmse_small = round(np.sqrt(mean_squared_error(y, lm.predict(X))))
print(rmse_small)
```

```python
# Using all data:
rmse_all = round(np.sqrt(mean_squared_error(y_all, lm.predict(X_all))))
print(rmse_all)
```

```python
# The percentage error of the model with the small dataset is:
str(round(rmse_small/np.mean(y) * 100)) + '%'
```

```python
# The percentage error of the model with the full dataset is:
str(round(rmse_all/np.mean(y_all) * 100)) + '%'
```

```python
# We will compute R square which measures the proportion
# of variability in Y that can be explained using X. 
# For the small dataset:
r2_small = round(lm.score(X, y), 3)
print("R squared of small_dataset: {}".format(r2_small))
```

```python
# For the large dataset:
r2_all = round(lm.score(X_all, y_all), 3)
print("R squared using all data: {}".format(r2_all)) 
```

```python
# Huge difference between the two datasets.
```
